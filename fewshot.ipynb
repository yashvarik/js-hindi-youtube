```python
from langchain_ollama import ChatOllama
```

```python
from langchain_core.prompts import HumanMessagePromptTemplate,AIMessagePromptTemplate,FewShotChatMessagePromptTemplate,ChatPromptTemplate
```

```python
chat=ChatOllama(
    model='llama3.1:8b')
```

```python
temp_h=''' can you explain me a {dish}'''
```

```python
temp_ai='''{response}'''
```

```python
mess_h=HumanMessagePromptTemplate.from_template(template=temp_h)
```

```python
mess_ai=AIMessagePromptTemplate.from_template(template=temp_ai)
```

```python
example_temp=ChatPromptTemplate.from_messages([mess_h,mess_ai])
```

```python
example=[{
    'dish':'vada',
    'response':''' oh,you want to know its is the the same dish enjoy in mumbai'''},
         {
             'dish':'dosa',
             'response':''' Are you living under the rock ! it is the one of the best indian dishes'''}]
```

```python
few_shot_tem=FewShotChatMessagePromptTemplate(examples=example,
                                              example_prompt=example_temp)
```

```python
chat_template=ChatPromptTemplate.from_messages([few_shot_tem,
                                               mess_h])
```

```python
chat_value=chat_template.invoke({'dish':'samosa'})
                                 
```

```python
chat_value
```

```python
response=chat.invoke(chat_value)
```

```python
for i in chat_value.messages:
    print(f'{i.type}:{i.content}\n')
```
