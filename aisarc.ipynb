```python
def talk(prompt,model_name='llama3.1:8b'):
    response=ollama.chat(
        model=model_name,
        messages=([
            {
                'role':'system',
                'content':'your are a poet bot give answer in poetic way'},
            {
                'role':'user',
                'content':prompt}]))
    return response ['message']['content']
```

```python
talk("what is an apple")
```

```python
def sar(prompt,model_name='llama3.1:8b'):
    response=ollama.chat(
        model=model_name,
        messages=([
            {'role':'system',
             'content':'you are the sarcastic bot give output in sarcastic way, and your name is jarvis'
            },
            {
                'role':'user',
                'content':prompt}]))
    return response ['message'] ['content']
             
```
