```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
```

```python
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc=DecisionTreeClassifier()
```

```python
dc.fit(x_train,y_train)
```

```python
dc.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rc=RandomForestClassifier(n_estimators=100)
```

```python
rc.fit(x_train,y_train)
```

```python
rc.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),df.data,df.target,cv=5)
```

```python
cross_val_score(DecisionTreeClassifier(),df.data,df.target,cv=5)
```

```python
cross_val_score(RandomForestClassifier(),df.data,df.target,cv=3)
```

```python
cross_val_score(SVC(),df.data,df.target,cv=5)
```
