```python
from langchain_community.document_loaders import Docx2txtLoader
```

```python
from langchain_text_splitters.markdown import MarkdownHeaderTextSplitter
```

```python
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
loader=Docx2txtLoader(r"C:\Users\yash\Downloads\Introduction_to_Data_and_Data_Science_2.docx")
```

```python
pages_doc=loader.load()
```

```python
md_split=MarkdownHeaderTextSplitter(
    headers_to_split_on=[("#","cousre title"),("##","lecture title")]
)
```

```python
md_splitter=md_split.split_text(pages_doc[0].page_content
                               )
```

```python
for i in range(len(md_splitter)):
    md_splitter[i].page_content=' '.join(md_splitter[i].page_content.split())
```

```python
char_split=CharacterTextSplitter(
    separator='',
        chunk_size=500,
    chunk_overlap=50
)
```

```python
char_splitted=char_split.split_documents(md_splitter)
```
