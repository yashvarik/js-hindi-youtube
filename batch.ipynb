```python
from langchain_ollama import ChatOllama
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
chat=ChatOllama(
    model='llama3.1:8b')
```

```python
chat_promt=ChatPromptTemplate.from_messages([('human',"i got a {pet}, it is of {breed} can you suggest the name")])
```

```python
chain =chat_promt|chat
```

```python
chain.batch([{'pet':'cat',
            'breed':'siberian'},{'pet':'dog',
                                'breed':'lab'}])
```
