```python
import pandas as pd
```

```python
import matplotlib.pylab as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv(r"C:\Users\yash\Downloads\income.csv")
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
km=KMeans(n_clusters=3)
y_predic=km.fit_predict(df[['Age','Income($)']])
```

```python
y_predic
```

```python
df['cluster']=y_predic
```

```python
df.head()
```

```python
df1=df[df.cluster==0]
```

```python
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
df1
```

```python
plt.scatter(df1['Age'],df1['Income($)'])
plt.scatter(df2['Age'],df2['Income($)'])
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
```

```python
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
```

```python
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'])
plt.scatter(df2['Age'],df2['Income($)'])
plt.scatter(df3['Age'],df3['Income($)'])
plt.xlabel('age')
plt.ylabel('income')
plt.legend('best')
plt.show()
```

```python
km=KMeans(n_clusters=3)
y_predic=km.fit_predict(df[['Age','Income($)']])
```

```python
y_predic
```

```python
df['cluster']=y_predic
```

```python
df.head()
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'])
plt.scatter(df2['Age'],df2['Income($)'])
plt.scatter(df3['Age'],df3['Income($)'])
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="red",marker='*')
plt.xlabel('age')
plt.ylabel('income')
plt.legend('best')
plt.show()
```
