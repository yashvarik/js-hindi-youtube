```python
import langchain_ollama
```

```python
from langchain_ollama import ChatOllama
from langchain_core.messages import SystemMessage,HumanMessage
```

```python
chat=ChatOllama(model='llama3.1:8b')
```

```python
human_m=HumanMessage(content=''' what is pokemon ''')
```

```python
system_m=SystemMessage(content=' you are a sarcastic bot give output in sarcastic way')
```

```python
response=chat.invoke([system_m,human_m])
```

```python
print(response.content)
```
