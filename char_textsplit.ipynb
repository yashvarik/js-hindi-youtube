```python
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
from langchain_community.document_loaders import Docx2txtLoader
```

```python
loader=Docx2txtLoader(r"C:\Users\yash\Downloads\Introduction_to_Data_and_Data_Science.docx")
```

```python
pages_doc=loader.load()
```

```python
pages_doc[0]
```

```python
for i in range (len(pages_doc)):
    pages_doc[i].page_content=' '.join(pages_doc[i].page_content.split())
```

```python
pages_doc[0]
```

```python
char_split=CharacterTextSplitter(separator='',
                                chunk_size=500,
                                chunk_overlap=0)
```

```python
pages_cut=char_split.split_documents(pages_doc)
```

```python
pages_cut[0]
```

```python
for i in range (len(pages_cut)):
    pages_cut[i].page_content= pages_cut[i].page_content+'\n'
```
