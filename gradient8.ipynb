```python
import numpy as np
def grad_des(x,y):
    m_curr = 0
    b_curr = 0
    iteration = 10
    n=(len(x))
    learn = 0.001
    for i in range(iteration):
        yp= m_curr * x +b_curr
        cost= (1/n) * sum((y-yp**2))
        dm = -(2/n) * sum (x*(y-yp))
        db = -(2/n) * sum (y-yp)
        m_curr = learn * dm
        b_curr = learn * db
        print(" m {} , b {} , cost {} , i {}" .format(m_curr,b_curr,cost,i))
        pass

       
        
    
   






x=np.array([2,3,4,6])
y=np.array([10,20,30,40,])





grad_des(x,y)
              
```
