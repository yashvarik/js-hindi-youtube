```python
review=("the movie was nice",
        "the movie was good",
        "yash said it was nice",
        "sheetal said it was ok",)
good_review=[]
pattern_find=r"good"
for string in review:
    if re.search(pattern_find,string):
      good_review.append(string)
    print(good_review)
    
```

```python
end_e=r"e$"
new_e=[]
for string in review:
    if re.search(end_e,string):
        new_e.append(string)
        print(new_e)
        
```

```python
bucket=("wheat",
         "jaggary",
         "rice",
         "fruit",
         "veg",)

pattern_find=r"(wheat|rice)ed"
mon_it=[]

for iteam in bucket:
    if re.search(pattern_find,iteam):
        mon_it.append(iteam)
        print(mon_it)
```
