outlier detection using  IQR

```python
hei=pd.read_csv(r"C:\Users\yash\Downloads\heights.csv")
```

```python
hei.head()
```

```python
q1=hei.height.quantile(0.25)
```

```python
q3=hei.height.quantile(0.75)
```

```python
IQR=q3 - q1
IQR
```

```python
lower_limit=q1-1.5*IQR
upper_limit=q3+1.5*IQR
```

```python
lower_limit,upper_limit
```

```python
hei[(hei.height<lower_limit) | (hei.height>upper_limit)]
```

```python
df5=hei[(hei.height>lower_limit) & (hei.height<upper_limit)]
```

```python
df5
```
